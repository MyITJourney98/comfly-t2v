```python
# Install pyngrok if not done
!pip install pyngrok

# Launch ComfyUI
import threading, time, subprocess
from pyngrok import ngrok
from IPython.display import HTML

def run_comfy():
    subprocess.run(["python", "main.py", "--listen", "0.0.0.0", "--port", "8188"])

threading.Thread(target=run_comfy).start()
time.sleep(10)

public_url = ngrok.connect(8188)
HTML(f"<b>✅ ComfyUI is running at:</b> <a href='{public_url}' target='_blank'>{public_url}</a>")
```
